## **This is a code to train perceptron weights**
We have an inital equation 3x1+4x2-10, and we want it to classify the point (1,1) correctly, with a learning rate of 0.1:

```python
import numpy as np
import pandas as pd

weights = np.array([3., 4., -10.])
learning_rate = 0.1
point_bias = np.array([1., 1., 1.])
counter = 0
Results = pd.DataFrame(columns=['cycle','linear_combination','weights0','weights1','bias'])

while True:
    linear_combination = weights[0] * point_bias[0] + weights[1] * point_bias[1] + weights[2]
    output = linear_combination >= 0
    counter += 1

    Results.loc[counter] = [counter, linear_combination, weights[0], weights[1], weights[2]]
    if output:
        print(f"Finish in {counter} cycles")
        break
    weights = weights + learning_rate * point_bias

Results
```
